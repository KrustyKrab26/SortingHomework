### Tạo 10 mảng, mỗi mảng 1 triệu số, bao gồm các số ngẫu nhiên

Yêu cầu:
- 1 mảng số thực có thứ tự tăng dần.
- 1 mảng số thực có thứ tự giảm dần.
- 8 mảng thứ tự ngẫu nhiên, bao gồm:
    - 4 mảng số thực.
    - 4 mảng số nguyên.

```python
import numpy as np
# Seed random
np.random.seed(42)
# Kích thước mảng
n = 1_000_000

# 1. Mảng số thực sắp xếp theo thứ tự tăng dần
arr_asc_float = np.sort(np.random.rand(n))

# 2. Mảng số thực sắp xếp theo thứ tự giảm dần
arr_desc_float = np.sort(np.random.rand(n))[::-1]

# 3. 8 mảng ngẫu nhiên
# 4 mảng số thực ngẫu nhiên
arr_rand_floats = [np.random.rand(n) for _ in range(4)]

# 4 mảng số nguyên ngẫu nhiên (giả sử trong khoảng 0 đến 1.000.000)
arr_rand_ints = [np.random.randint(0, n, n) for _ in range(4)]
```

```python
def quick_sort(arr):
    # Trường hợp cơ bản: mảng rỗng hoặc chỉ có một phần tử
    if len(arr) <= 1:
        return arr
    else:
        # Chọn phần tử pivot (ở đây chọn phần tử ở giữa)
        pivot = arr[len(arr) // 2]
        
        # Các phần tử nhỏ hơn pivot
        left = [x for x in arr if x < pivot]
        
        # Các phần tử bằng pivot
        middle = [x for x in arr if x == pivot]
        
        # Các phần tử lớn hơn pivot
        right = [x for x in arr if x > pivot]
        
        # Đệ quy sắp xếp các phần tử bên trái và bên phải rồi nối lại
        return quick_sort(left) + middle + quick_sort(right)
```

```python
# Đo thời gian chạy quick_sort trên mảng số thực tăng dần
%time sorted_arr = quick_sort(arr_asc_float)
```

```python
# Đo thời gian chạy quick_sort trên mảng số thực giảm dần
%time sorted_arr = quick_sort(arr_desc_float)
```

```python
# Đo thời gian chạy quick_sort trên 4 mảng số thực ngẫu nhiên
for i, arr in enumerate(arr_rand_floats):
    %time sorted_arr = quick_sort(arr)
```

```python
# Đo thời gian chạy quick_sort trên 4 mảng số nguyên ngẫu nhiên
for i, arr in enumerate(arr_rand_ints):
    %time sorted_arr = quick_sort(arr)
```
